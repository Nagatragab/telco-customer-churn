```python
from sklearn.model_selection import train_test_split

# --- Preparatory Steps (ensure this is run) ---

# Let's assume df is the original DataFrame after cleaning
# 1. Define the independent variable X and the target variable y
X = df.drop(columns=['Churn', 'customerID'])
y = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0) # <-- Here we define 'y'

# 2. Encode categorical variables
X_encoded = pd.get_dummies(X, drop_first=True)


# --- The code you were trying to run (will work now) ---

# 3. Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, 
    y, 
    test_size=0.2, 
    random_state=42, 
    stratify=y
)

# 4. Print the shapes of the sets to confirm the process was successful
print("Data split successfully!")
print("----------- Data shapes after splitting -----------")
print("Training data shape (X_train):", X_train.shape)
print("Test data shape (X_test):", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. Build and Train the Logistic Regression Model ---

# Initialize the model
log_model = LogisticRegression(max_iter=1000, random_state=42)

# Train the model using the training data
print("Training the Logistic Regression model...")
log_model.fit(X_train, y_train)
print("...Training complete!")

# Make predictions on the test data (which the model has not seen before)
y_pred_log = log_model.predict(X_test)

# Evaluate the model's performance
print("\n--------- [ Logistic Regression Model Results ] ---------")
print(f"Accuracy: {accuracy_score(y_test, y_pred_log):.4f}")
print("\nClassification Report:")
# We print the classification report which shows Precision, Recall, F-Score
print(classification_report(y_test, y_pred_log, target_names=['No Churn', 'Churn']))


# --- 2. Build and Train the Random Forest Model ---

# Initialize the model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
print("\n\nTraining the Random Forest model...")
rf_model.fit(X_train, y_train)
print("...Training complete!")

# Make predictions on the test data
y_pred_rf = rf_model.predict(X_test)

# Evaluate the model's performance
print("\n--------- [ Random Forest Model Results ] ---------")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, target_names=['No Churn', 'Churn']))


# --- 3. Plotting the Confusion Matrix for the best model ---
# (Random Forest is usually the best)

print("\n\n--- Confusion Matrix for the Random Forest Model ---")
cm = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Random Forest')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# 1. Define the parameter grid we want to try
# We've chosen a small grid to run quickly, you can expand it later
param_grid = {
    'n_estimators': [100, 150],      # Number of trees in the forest
    'max_depth': [10, 20, None],     # Maximum depth of the tree (None means no limit)
    'min_samples_leaf': [1, 2],      # Minimum number of samples in a leaf node
    'criterion': ['gini', 'entropy'] # The quality measure of a split
}

# 2. Create the GridSearchCV instance
# estimator is a base Random Forest model
# scoring='recall' defines our main objective
# cv=3 means using 3-fold cross-validation
# n_jobs=-1 uses all CPU cores to speed up the process
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42), 
    param_grid=param_grid, 
    scoring='recall', 
    cv=3, 
    n_jobs=-1,
    verbose=2 # To display the process details during execution
)

# 3. Start the search and training process on the balanced data
print("Searching for the best parameters for the Random Forest model...")
# We use X_train_resampled and y_train_resampled
grid_search.fit(X_train_resampled, y_train_resampled)
print("...Search finished!")

# 4. Print the best combination found
print("\nBest parameters found:")
print(grid_search.best_params_)

# 5. Evaluate the best model on the test data
best_rf_model = grid_search.best_estimator_
y_pred_best_rf = best_rf_model.predict(X_test)

print("\n--------- [ Optimized Random Forest Model Results ] ---------")
print(classification_report(y_test, y_pred_best_rf, target_names=['No Churn', 'Churn']))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# --- 1. Feature Engineering ---
# Let's assume df is the original clean DataFrame
print("Creating a new feature...")

# For safety, replace any tenure=0 value with 1 to avoid division by zero
df['tenure'].replace(0, 1, inplace=True)
df['charge_per_tenure'] = df['MonthlyCharges'] / df['tenure']
# Handle any infinite values that might result
df.replace([np.inf, -np.inf], 0, inplace=True)

print("...Feature created successfully!")


# --- 2. Full Data Reprocessing ---
print("\nReprocessing the data with the new feature...")

# Redefine X and y
X_new = df.drop(columns=['Churn', 'customerID'])
y_new = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

# Apply One-Hot Encoding
X_new_encoded = pd.get_dummies(X_new, drop_first=True)

# Apply Scaling (adding the new feature to the list)
numerical_cols_new = ['tenure', 'MonthlyCharges', 'TotalCharges', 'charge_per_tenure']
scaler = StandardScaler()
X_new_encoded[numerical_cols_new] = scaler.fit_transform(X_new_encoded[numerical_cols_new])

# Split the data
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(
    X_new_encoded, y_new, test_size=0.2, random_state=42, stratify=y_new
)

# Apply SMOTE to the new training data
smote = SMOTE(random_state=42)
X_train_resampled_new, y_train_resampled_new = smote.fit_resample(X_train_new, y_train_new)

print("...Processing complete!")


# --- 3. Retrain and Evaluate the Champion Model ---
print("\nRetraining the best model on the new data...")

# Use the best parameters we found previously
best_log_model_final = LogisticRegression(C=0.01, penalty='l1', solver='liblinear', random_state=42, max_iter=1000)
best_log_model_final.fit(X_train_resampled_new, y_train_resampled_new)

print("...Training complete!")

# Final Evaluation
y_pred_final = best_log_model_final.predict(X_test_new)

print("\n--------- [ Final Results after Feature Engineering ] ---------")
print(classification_report(y_test_new, y_pred_final, target_names=['No Churn', 'Churn']))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# --- 1. Feature Engineering ---
# Let's assume df is the original clean DataFrame
print("Creating a new feature...")
df['tenure'].replace(0, 1, inplace=True) # to avoid division by zero
df['charge_per_tenure'] = df['MonthlyCharges'] / df['tenure']
df.replace([np.inf, -np.inf], 0, inplace=True)
print("...Feature created successfully!")


# --- 2. Full Data Reprocessing ---
print("\nReprocessing the data with the new feature...")
X_new = df.drop(columns=['Churn', 'customerID'])
y_new = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)
X_new_encoded = pd.get_dummies(X_new, drop_first=True)
numerical_cols_new = ['tenure', 'MonthlyCharges', 'TotalCharges', 'charge_per_tenure']
scaler = StandardScaler()
X_new_encoded[numerical_cols_new] = scaler.fit_transform(X_new_encoded[numerical_cols_new])
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(
    X_new_encoded, y_new, test_size=0.2, random_state=42, stratify=y_new
)
print("...Processing complete!")


# --- 3. Balancing the Training Data ---
print("\nBalancing the training data using SMOTE...")
smote = SMOTE(random_state=42)
X_train_resampled_new, y_train_resampled_new = smote.fit_resample(X_train_new, y_train_new)
print("...Balancing complete!")


# --- 4. Hyperparameter Tuning and Training ---
print("\nSearching for the best parameters for the Random Forest model...")
param_grid_rf = {
    'n_estimators': [100, 150],
    'max_depth': [10, 20, None],
    'min_samples_leaf': [1, 2],
    'criterion': ['gini', 'entropy']
}
grid_search_rf_final = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42), 
    param_grid=param_grid_rf, 
    scoring='recall', 
    cv=3, 
    n_jobs=-1,
    verbose=2
)
grid_search_rf_final.fit(X_train_resampled_new, y_train_resampled_new)
print("...Search finished!")


# --- 5. Final Evaluation ---
print("\nBest parameters found:")
print(grid_search_rf_final.best_params_)
best_rf_model_final = grid_search_rf_final.best_estimator_
y_pred_rf_final = best_rf_model_final.predict(X_test_new)

print("\n--------- [ Final Results of the Strongest Model ] ---------")
print(classification_report(y_test_new, y_pred_rf_final, target_names=['No Churn', 'Churn']))
```
